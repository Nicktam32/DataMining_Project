```python
import pandas as pd
import numpy as np

# 读取数据
df_population = pd.read_csv("population.csv")
df = pd.read_csv("solid-waste-disposal-quantity-by-category-en-2022.csv")
df_gdp = pd.read_csv("GDP.csv")
```

```python
df.head()
```

```python
# 合并两个数据框
merged_df = pd.merge(df, df_population, how='left', left_on='year', right_on='Year')
```

```python
merged_df.head()
```

```python
# Filter df_gdp to include only years from 2009 to 2022
df_gdp_filtered = df_gdp[(df_gdp['year'] >= 2009) & (df_gdp['year'] <= 2022)]

# Convert the 'year' column data types to match (if necessary)
# For example, if 'year' is an object in merged_df and int in df_gdp_filtered:
merged_df['year'] = merged_df['year'].astype(int)

# Merge the dataframes
merged_df = pd.merge(merged_df, df_gdp_filtered, how='left', left_on='year', right_on='year')
```

```python
merged_df.head()
```

```python
# 按廢物類別劃分的固體廢物棄置總量（2009年至2022年）（英文版）
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 计算增长率
merged_df['growth'] = merged_df.groupby('waste_cat_en')['disposal_q'].pct_change()

# 绘制增长率图
plt.figure(figsize=(12, 6))
plt.plot(merged_df['growth'], label='Municipal solid waste')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.title('Waste Growth Rates')
plt.show()

# 时间序列预测
model = ARIMA(merged_df['disposal_q'], order=(1, 1, 1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=1)
```

```python
import seaborn as sns
# 选择需要绘制的列
cols = ['waste_cat_en', 'disposal_q', 'All age group population', 'growth', "GDP"]
data = merged_df[cols]

# 对'waste_cat_en'列进行虚拟变量编码
waste_dummies = pd.get_dummies(data['waste_cat_en'], prefix='waste')
data = pd.concat([data, waste_dummies], axis=1)
data = data.drop('waste_cat_en', axis=1)

# 计算相关性矩阵
corr = data.corr()

# 绘制热力图
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()

# 计算协方差矩阵
cov_matrix = data.cov()

# 打印协方差矩阵
print("协方差矩阵:")
print(cov_matrix)
```

```python
# 根据给出的协方差矩阵,我们可以对变量之间的相关性做出以下分析:

# 1. **废物处理量(disposal_q)与其他变量的关系**:
#    - 废物处理量与人口总数存在正相关关系,相关系数为 1.635080e+07。
#    - 废物处理量与GDP存在较强的正相关关系,相关系数为 4.162486e+10。
#    - 废物处理量与城市固体废物(Municipal solid waste)呈现正相关,与建筑垃圾(Overall construction waste)呈现负相关,与特殊废物(Special waste)呈现负相关。

# 2. **不同类型废物之间的关系**:
#    - 城市固体废物与建筑垃圾、特殊废物之间存在负相关关系,相关系数约为-0.114。
#    - 建筑垃圾与特殊废物之间也存在负相关关系,相关系数约为-0.114。

# 3. **人口与经济数据的关系**:
#    - 人口总数与GDP呈现正相关关系,相关系数为 6.728388e+07。
#    - 人口增长率(growth)与GDP呈现负相关关系,相关系数为-1925.054。

# 4. **其他发现**:
#    - 人口增长率与废物处理量呈现正相关关系,相关系数为 43059.342。
#    - 人口增长率与不同类型废物之间存在微弱的正相关或负相关关系。

# 基于这些发现,我们可以得出以下一些初步结论:

# - 废物处理量与人口总数和经济发展水平(GDP)密切相关,可能是主要的影响因素。
# - 不同类型的废物之间存在一定的替代关系或互斥关系,增加一种废物可能会导致另一种减少。
# - 人口增长率对废物处理量的影响较小,但仍存在一定的正相关关系。

# 因此,在构建回归模型时,我们可以优先考虑人口总数、GDP以及不同类型的废物作为自变量,尝试捕捉它们与废物处理量之间的线性或非线性关系。同时,也可以探索引入人口增长率等其他可能的影响因素,以期获得更好的预测效果。
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# 准备数据
X = merged_df[['year', 'All age group population', 'GDP']]
y = merged_df['disposal_q']

# 训练模型
model = LinearRegression()
model.fit(X, y)

# 评估模型
print('模型系数:', model.coef_)
print('模型截距:', model.intercept_)
print('模型R平方:', model.score(X, y))
```

```python
import torch
import torch.nn as nn

# 定义神经网络模型
class WasteRegressionModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(WasteRegressionModel, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

# 准备数据
X = merged_df[['GDP', 'All age group population', 'year']]
y = merged_df['disposal_q']

# 将数据转换为PyTorch张量
X = torch.tensor(X.values, dtype=torch.float32)
y = torch.tensor(y.values, dtype=torch.float32).unsqueeze(1)

# 定义模型超参数
input_size = X.shape[1]
hidden_size = 64
output_size = 1

# 实例化模型
model = WasteRegressionModel(input_size, hidden_size, output_size)

# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# 训练模型
num_epochs = 1000
for epoch in range(num_epochs):
    # 前向传播
    outputs = model(X)
    loss = criterion(outputs, y)

    # 反向传播和优化
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```
